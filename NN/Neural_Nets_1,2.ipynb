# Single Neural Network - One Neuron

# Input Layer
    input_layer = [x1, x2, x3]
    weights = [w1, w2, w3]
    bias = b
    
# Hidden Layer
    Hidden_layer = input_layer * weights
    => x1*w1 + x2*w2 + x3*w3

# Output Layer
    output = Hidden_layer + b
### O = I[0]*w[0] + I[1]*w[1] + I[2]*w[2] + b

```python
input_layer = [1, 2, 3]
weights = [1.0, 2.0, 3.0]
bias = 3

hidden_layer = sum([i*j for i,j in zip(input_layer, weights)])
print(hidden_layer)

Output_layer = hidden_layer + bias
print("{0}".format(Output_layer))
```

```python
# FUNCTIONAL APPROACH
def NN_fn(in, wts, bias):
    hidden_layer = sum([i*j for i,j in zip(in, wts)])
    print(hidden_layer)
    
    Out = hidden_layer + bias
    return Out
```

```python
# Input
I = [78, 183]

# weights
wts_1 = [2, 6] # weights for neuron 1
wts_2 = [3, 9] # weights for neuron 2
B1 = 4
B2 = 5

# Hidden
H1 = I[0]*wts_1[0] + I[1]*wts_1[1] + B1
H2 = I[0]*wts_2[0] + I[1]*wts_2[1] + B2

# Output
O= H1 + H2


print(O)
```
